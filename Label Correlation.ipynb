```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_raw = pd.read_csv("./kag_risk_factors_cervical_cancer.csv")
```

```python
df_raw.head()
```

```python
# Replace '?' with 'NaN' characters
df_raw_na = df_raw.replace('?', np.nan)
df_raw_na.isnull().sum()
```

```python
df_labels = df_raw.iloc[:,-4:]

corr = df_labels.corr()
ax = plt.axes()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)

ax.set_title('Label Correlation Matrix', fontsize = 16)
plt.show()
```
